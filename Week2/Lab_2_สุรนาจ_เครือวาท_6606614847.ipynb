```python
import folium
from folium import plugins
from folium.plugins import HeatMap, MiniMap, Fullscreen

# =========================
# คำนวณจุดกึ่งกลาง Tokyo
# =========================
center = [boundary.geometry.centroid.y.mean(),
          boundary.geometry.centroid.x.mean()]

# =========================
# สร้างแผนที่หลัก
# =========================
m = folium.Map(
    location=center,
    zoom_start=10,
    tiles=None,
    control_scale=True  # Scale bar
)

# =========================
# Basemap Options
# =========================
folium.TileLayer("OpenStreetMap", name="OpenStreetMap").add_to(m)
folium.TileLayer("CartoDB positron", name="Light Map").add_to(m)

folium.TileLayer(
    tiles="https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}",
    attr="Esri",
    name="Satellite",
).add_to(m)

# =========================
# Tokyo Boundary
# =========================
folium.GeoJson(
    boundary,
    name="Tokyo Boundary",
    style_function=lambda x: {
        "fillColor": "none",
        "color": "blue",
        "weight": 2
    }
).add_to(m)

# =========================
# Air Monitoring Stations
# =========================
station_group = folium.FeatureGroup(name="Air Monitoring Stations")

for idx, row in stations.iterrows():
    
    pm25_value = row.get("pm25", "No Data")

    folium.CircleMarker(
        location=[row.geometry.y, row.geometry.x],
        radius=7,
        color="red",
        fill=True,
        fill_color="yellow",
        fill_opacity=0.8,
        popup=f"""
        <b>Air Quality Monitoring Station</b><br>
        City: Tokyo<br>
        Name: {row.get("name","Station")}<br>
        PM2.5: {pm25_value}
        """
    ).add_to(station_group)

station_group.add_to(m)

# =========================
# Heatmap Layer
# =========================
heat_data = [[row.geometry.y, row.geometry.x] 
             for idx, row in stations.iterrows()]

HeatMap(
    heat_data,
    name="Station Density Heatmap",
    radius=25,
    blur=20
).add_to(m)

# =========================
# MiniMap
# =========================
minimap = MiniMap(toggle_display=True)
m.add_child(minimap)

# =========================
# Fullscreen Button
# =========================
Fullscreen(position='topright').add_to(m)

# =========================
# Title
# =========================
title_html = """
     <h3 align="center" style="font-size:20px">
     <b>Air Quality Monitoring Stations & Density Analysis<br>
     Tokyo, Japan</b>
     </h3>
     """
m.get_root().html.add_child(folium.Element(title_html))

# =========================
# Legend
# =========================
legend_html = '''
<div style="
position: fixed; 
bottom: 50px; left: 50px; width: 260px; height: 166px; 
background-color: white; 
border:2px solid grey; 
z-index:9999; 
font-size:14px;
padding: 10px;
">
<b>Legend</b><br>
🔵 Blue Line = Tokyo Boundary<br>
🟡🟥 Circle = Air Monitoring Station<br>
🔥 Heatmap = Station Density<br>
📏 Scale bar (bottom left)<br>
🧭 Mini map (corner)<br>
📸 Fullscreen button (top right)
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# =========================
# Layer Control
# =========================
folium.LayerControl(collapsed=False).add_to(m)

m
```

```python
import geopandas as gpd
import rasterio
from rasterio.mask import mask
from rasterio.plot import show
from shapely.geometry import mapping
import matplotlib.pyplot as plt
```

```python
# โหลด boundary
boundary = gpd.read_file("/home/lenovo/Documents/GotWork/337/code/Tokyo/")

# เปิด raster
src = rasterio.open("/home/lenovo/Documents/GotWork/337/R10m/T54SUE_20260116T012919_B04_10m.jp2")

# แปลง CRS ให้ตรงกัน
boundary = boundary.to_crs(src.crs)
```

```python
show(src, title="Sentinel-2 B04 (Tokyo2026)")
```

```python
print("==== Raster Metadata ====")
print("Driver:", src.driver)
print("Width (pixels):", src.width)
print("Height (pixels):", src.height)
print("Number of Bands:", src.count)
print("CRS:", src.crs)
print("Resolution (pixel size):", src.res)
print("Bounds:", src.bounds)
print("Data Type:", src.dtypes)
```

```python
# แปลง geometry เป็น GeoJSON format
geoms = [mapping(geom) for geom in boundary.geometry]

# ทำ masking
out_image, out_transform = mask(src, geoms, crop=True)

# คัดลอก metadata เดิม
out_meta = src.meta.copy()

# ปรับ metadata ใหม่
out_meta.update({
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform
})

# บันทึกไฟล์ใหม่
with rasterio.open("B04_clipped.tif", "w", **out_meta) as dest:
    dest.write(out_image)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(out_image[0], cmap='rainbow')
plt.title("Clipped Red Band - Tokyo")
plt.colorbar()
plt.show()
```

```python
import rasterio
import numpy as np

with rasterio.open("B08_clipped.tif") as nir_src:
    nir = nir_src.read(1).astype("float32")
    profile = nir_src.profile

with rasterio.open("B04_clipped.tif") as red_src:
    red = red_src.read(1).astype("float32")

np.seterr(divide='ignore', invalid='ignore')

ndvi = (nir - red) / (nir + red)
ndvi[np.isnan(ndvi)] = -9999

# 🔥 สำคัญมาก
profile.update(
    driver="GTiff",      # เปลี่ยน driver
    dtype=rasterio.float32,
    count=1,
    nodata=-9999
)

with rasterio.open("NDVI.tif", "w", **profile) as dst:
    dst.write(ndvi.astype(rasterio.float32), 1)

print("NDVI saved successfully as GeoTIFF")
```

```python
with rasterio.open("NDVI.tif") as src:
    ndvi_data = src.read(1)
    ndvi_data = np.where(ndvi_data == -9999, np.nan, ndvi_data)

plt.figure(figsize=(8,6))
img = plt.imshow(ndvi_data, cmap="RdYlGn", vmin=-1, vmax=1)
plt.colorbar(img, label="NDVI")
plt.title("NDVI Map - Tokyo")
plt.xlabel("Column")
plt.ylabel("Row")
plt.show()
```

```python
with rasterio.open("NDVI.tif") as src:
    ndvi_data = src.read(1)
    ndvi_data = np.where(ndvi_data == -9999, np.nan, ndvi_data)
    
    extent = [
        src.bounds.left,
        src.bounds.right,
        src.bounds.bottom,
        src.bounds.top
    ]

plt.figure(figsize=(8,6))
img = plt.imshow(ndvi_data, cmap="RdYlGn", extent=extent)
plt.colorbar(img, label="NDVI")
plt.title("NDVI Map with Coordinates - Tokyo")
plt.xlabel("Longitude / Easting")
plt.ylabel("Latitude / Northing")
plt.show()
```

```python
valid_pixels = ndvi_data[~np.isnan(ndvi_data)]

plt.figure(figsize=(6,5))
plt.hist(valid_pixels.flatten(), bins=40)
plt.title("NDVI Distribution")
plt.xlabel("NDVI Value")
plt.ylabel("Frequency")
plt.show()
```
